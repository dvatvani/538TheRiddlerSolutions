# Riddler Express - 2020-04-10

This is a solution to the Riddler Express at FiveThirtyEight, dated 10th April 2020. The original riddle can be found on their website [here](https://fivethirtyeight.com/features/can-you-catch-the-free-t-shirt/). The problem for this week is formulated as follows:

-----

During a break at a music festival, the crew is launching T-shirts into the audience using a T-shirt cannon. And you’re in luck — your seat happens to be in the line of flight for one of the T-shirts! In other words, if the cannon is strong enough and the shirt is launched at the right angle, it will land in your arms.

The rows of seats in the audience are all on the same level (i.e., there is no incline), they are numbered 1, 2, 3, etc., and the T-shirts are being launched from directly in front of Row 1. Assume also that there is no air resistance (yes, I know, that’s a big assumption). You also happen to know quite a bit about the particular model of T-shirt cannon being used — with no air resistance, it can launch T-shirts to the very back of Row 100 in the audience, but no farther.

The crew member aiming in your direction is still figuring out the angle for the launch, which you figure will be a random angle between zero degrees (straight at the unfortunate person seated in Row 1) and 90 degrees (straight up). Which row should you be sitting in to maximize your chances of nabbing the T-shirt?

-----

## Solution

This is clearly a question driven by the mechanics of projectile motion. If we sketch a diagram of the path of the projectile and split it's initial speed into horizontal and vertical components, it will allow us to determine the landing horizontal distance travelled at any given firing angle $\theta$

<img src="https://d2jmvrsizmvf4x.cloudfront.net/EpXDtQVfRAiohHNVeN01_2d+projectile+motion.jpg" width="400" height="400">
<center>Image from <a href="https://socratic.org/questions/58a6e62ab72cff6247458296">Socratic.org</a></center>

\begin{equation*}
u_x = u \times cos(\theta)
\end{equation*}
\begin{equation*}
u_y = u \times sin(\theta)
\end{equation*}

Newton's laws of motion will be handy for this problem, particularly the equation 

\begin{equation*}
s = ut + \frac{1}{2} a t^2
\end{equation*}

Considering the vertical movement, we know that the point at which the t-shirt lands will be the same height level as the t-shirt projectile was launched from i.e. net vertical displacement $s_y = 0$. We can use this to deduce the flight time of the projectile until it lands.

\begin{equation*}
s_y = u_yt + 1/2 a_y t^2 = 0
\end{equation*}

where $a_y$ is the downwards acceleration from gravity

From which, if we apply the quadratic formula ($x=\frac{-b\pm\sqrt{b^2-4ac}}{2a}$) with $x=t$, $a=u_y$, $b=\frac{1}{2}a_y$ and $c=0$, we can obtain the obtain the expression for projectile flight time 

\begin{equation*}
t=\frac{-u_y\pm{u_y}}{a}
\end{equation*}

Therefore, the solutions for when the t-shirt is at ground level are $t=0$ (the starting point of the parabolic trajectory when the t-shirt just leaves the cannon), and $t=-2u_y/a$, which is when the t-shirt projectile lands in the crowd

(It's worth remembering that the value of $a$ is negative in the equation above, since it's the acceleration due to gravity, acting in a direction opposite to the vertical component of the initial t-shirt projection velocity), commonly apploximated as -9.81

At this time t, the horizontal distance covered by the projectile is 

\begin{equation*}
s_x = u_xt
\end{equation*}

(from $s = ut + \frac{1}{2} a t^2$ with $a=0$ since there is no source of horizontal acceleration once the cannon fires)

Substituting in th expression for $t$ from earlier, we get

\begin{equation*}
s_x = \frac{-2u_xu_y}{a_y}
\end{equation*}

Or, remembering that $u_x = u \times cos(\theta)$ and $u_y = u \times sin(\theta)$,

\begin{equation*}
s_x = \frac{-2u^2 sin(\theta) cos(\theta)}{a_y}
\end{equation*}

where $a_y = -9.81$

However, since we know that the speed of the cannon is fixed and normalised such that it hits 100 rows far at max reach, the key feature of the above equation for us is that 
\begin{equation*}
s_x \propto sin(\theta) cos(\theta)
\end{equation*}

This confirms the fact that the furthest distance achieved by a projectile on flat ground is if it's launched at 45 degrees up from the plane of the ground, but also gives shows us the exact relationship between firing angle and distance achieved

```python
# A quick way of looking at the likely distances where the t-shirt will land if the angle 
# is randomly chosen between 0 and 90 is to look at the histogram of s_x for evenly sampled theta

theta = np.linspace(0,np.pi/2,10000)
s_x = np.sin(theta) * np.cos(theta)

# Histogram of vertical distance covered on evenly scanning theta from 0 to 90 degrees
plt.hist(s_x*200, bins=100, density=True);
plt.xlabel('Row number')
plt.ylabel('Likelihood of receiving free t-shirt')
plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x,y:f"{100*x:.0f}%"))
```

```python
# To demonstrate this more clearly, let's amimate the trajectory of the projectile from theta=0 to theta=90 degrees

theta_values = np.linspace(0,.5*np.pi,200)
for i, theta in tqdm(enumerate(theta_values)):
    plt.figure()
    
    g=-1
    u=10
    u_x = u*np.cos(theta)
    u_y = u*np.sin(theta)

    max_t = -2*u*np.sin(theta)/g
    t = np.linspace(0,max_t,100)

    s_x = u_x * t
    s_y = u_y * t + .5 * g * t**2

    
    # Plot histogram so far
    hist_theta = np.linspace(0,theta,10000)
    hist_s_x = np.sin(hist_theta) * np.cos(hist_theta)
    plt.hist(hist_s_x*200, bins=100, range=(0,100), normed=True, alpha=.3);
    plt.xlabel('Row number')
    plt.ylabel('Cumulative Likelihood of\nreceiving free t-shirt so far')
    plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x,y:f"{100*x:.1f}%"))
    left, bottom, width, height = plt.gca().get_position().bounds
    plt.gca().set_position([left+.1, bottom, width-.1, height] )
    
    # Plot trajectory of projectile
    plt.twinx()
    plt.plot(s_x, s_y)
    plt.xlim(-.5,100.5)
    plt.ylim(0,plt.ylim()[1])
    plt.yticks([])
    left, bottom, width, height = plt.gca().get_position().bounds
    plt.gca().set_position([left+.1, bottom, width-.1, height] )

    plt.savefig(f"temp/{i}.png")
    
    plt.close('all')

```
