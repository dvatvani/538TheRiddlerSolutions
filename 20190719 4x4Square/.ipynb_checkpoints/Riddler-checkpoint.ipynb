```python
import pandas as pd
import numpy as np
import itertools
from tqdm.autonotebook import tqdm
```

```python
# 65,536 unique combinations to colour in the grid (2^16)
```

```python
large_grid_size = 4
```

```python
solutions = []
all_possible_n_bit_combinations = itertools.product(*[[0,1]]*(large_grid_size**2))
for an_n_bit_combination in all_possible_n_bit_combinations:
    grid = np.array(an_n_bit_combination).reshape(large_grid_size,large_grid_size)
    all_2x2_subgrids = [grid[x:x+2,y:y+2] for x in range(large_grid_size-1) for y in range(large_grid_size-1)]
    if all([(pair[0] != pair[1]).any() for pair in itertools.combinations(all_2x2_subgrids, 2)]):
        solutions.append(grid)
```

```python
len(solutions)
```

```python
solutions[1][::-1]
```

```python
# Remove symmetrically related_solutions 

unique_solutions = []

for solution in tqdm(solutions):
    # Add initial figure and all 90 degree rotations
    symmetry_equivalent_solutions = [solution,
                                     np.rot90(solution),
                                     np.rot90(solution,2),
                                     np.rot90(solution,3)]
    
    # Add mirror image for each existing variant
    symmetry_equivalent_solutions += [x[::-1] for x in symmetry_equivalent_solutions]
    
    # Add 0-1 bit flip for each existing variant
    symmetry_equivalent_solutions += [1-x for x in symmetry_equivalent_solutions]
    
    if any([(new_sol == x).all() for new_sol in symmetry_equivalent_solutions for x in unique_solutions]):
        continue
    else:
        unique_solutions.append(solution)
    
```
